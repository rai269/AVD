### **3. Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, '% Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

### **5. Penanganan data**

**Missing Values**

```python
df['description'].fillna("No description available", inplace=True)
df['age_certification'].fillna(df['age_certification'].mode()[0], inplace=True)
df['seasons'].fillna(df['seasons'].median(), inplace=True)
df['imdb_score'].fillna(df['imdb_score'].median(), inplace=True)
df['imdb_votes'].fillna(df['imdb_votes'].median(), inplace=True)
df['tmdb_popularity'].fillna(df['tmdb_popularity'].median(), inplace=True)
df['tmdb_score'].fillna(df['tmdb_score'].median(), inplace=True)
df['imdb_id'].fillna("Unknown", inplace=True)
```

```python
print((df.isna().sum() / len(df)) * 100)
```

data yang hilang sudah tidak ada

**Outliers**

```python
columns_to_impute = ["release_year", "runtime", "seasons", "imdb_score", "imdb_votes", "tmdb_popularity", "tmdb_score"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, '% Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

## **Jumlah Judul Per Tahun**

```python
anime_per_year = df.groupby('release_year').size()
plt.figure(figsize=(10, 5))
sns.lineplot(x=anime_per_year.index, y=anime_per_year.values, marker='o', color='blue')
plt.title('Tren Jumlah Anime Per Tahun', fontsize=14)
plt.xlabel('Tahun Rilis', fontsize=12)
plt.ylabel('Jumlah Anime', fontsize=12)
plt.grid()
plt.show()
```

**INSIGHT**

Grafik menunjukkan bahwa jumlah judul anime yang dirilis meningkat drastis sejak tahun 2000-an, dengan puncaknya sekitar tahun 2018-2020. Beberapa faktor yang mungkin memengaruhi peningkatan ini:



1. **Popularitas Global** - Anime semakin diterima secara internasional berkat platform streaming seperti Netflix dan Crunchyroll.


2. **Kemajuan Teknologi** - Produksi anime menjadi lebih efisien dengan teknologi digital dibandingkan animasi manual.


3. **Permintaan Pasar** - Bertambahnya penggemar anime mendorong studio untuk memproduksi lebih banyak konten.


Namun, terlihat penurunan setelah 2020, yang kemungkinan disebabkan oleh pandemi COVID-19 yang menghambat produksi.

## **IMDB Score vs TMDB Score**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='imdb_score', y='tmdb_score', color='blue', alpha=0.5)
plt.title("Hubungan IMDb Score vs TMDb Score")
plt.xlabel("IMDb Score")
plt.ylabel("TMDb Score")
plt.grid()
plt.show()
```

### **Korelasi**

```python
df['imdb_score'] = pd.to_numeric(df['imdb_score'], errors='coerce')
df['tmdb_score'] = pd.to_numeric(df['tmdb_score'], errors='coerce')
correlation = df[['imdb_score', 'tmdb_score']].corr(method='pearson')
print(correlation)
```

**INSIGHT**


Dari scatter plot, terlihat bahwa skor IMDb dan TMDb memiliki korelasi sebesar **0.44**, yang menunjukkan **hubungan positif lemah hingga sedang**. Artinya:  

* **Skor di IMDb dan TMDb cenderung sejalan**, tetapi tidak sepenuhnya identik.


* **Variasi skor menunjukkan perbedaan preferensi pengguna**, di mana salah satu platform mungkin lebih kritis dalam menilai.


* **Sebaran data menunjukkan adanya outlier**, di mana beberapa film mendapat skor tinggi di satu platform tetapi lebih rendah di platform lainnya.



Secara keseluruhan, meskipun ada hubungan antara IMDb dan TMDb score, **penilaian di kedua platform memiliki perbedaan yang cukup signifikan**.

## **TOP 10 GENRE ANIME**

**Memisahkan Genre**

```python
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])
df_exploded = df.explode('genres')
```

**Visualisasi data**

```python
plt.figure(figsize=(8, 5))
df_exploded['genres'].value_counts().nlargest(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Genre Anime')
plt.xlabel('Genre')
plt.ylabel('Jumlah Anime')
plt.xticks(rotation=45)
plt.show()
```

**INSIGHT**  

Dari bar chart **Top 10 Genre Anime**, terlihat bahwa beberapa genre mendominasi jumlah produksi anime. Artinya:  

- **Genre Populer:** Genre dengan jumlah anime terbanyak menunjukkan tren yang digemari oleh industri dan penonton, seperti **action, adventure, dan fantasy**, yang sering hadir dalam anime populer.  

- **Permintaan Pasar:** Banyaknya anime dalam suatu genre mengindikasikan bahwa genre tersebut memiliki basis penggemar yang besar, sehingga studio lebih sering memproduksi anime dengan tema tersebut.  

- **Diversifikasi Konten:** Genre seperti **comedy, romance, dan sci-fi** juga masuk dalam daftar, menunjukkan variasi preferensi penonton terhadap cerita yang ringan hingga kompleks.  

- **Pengaruh Industri:** Studio besar mungkin lebih fokus pada genre-genre tertentu yang memiliki rekam jejak kesuksesan tinggi, sementara genre niche memiliki produksi yang lebih sedikit.  

Secara keseluruhan, tren genre anime menunjukkan bahwa industri anime beradaptasi dengan permintaan pasar, di mana genre action, adventure, dan fantasy tetap menjadi favorit utama.
